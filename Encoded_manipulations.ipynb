```python
import os
import numpy as np
import pickle
from PIL import Image
from __future__ import division

a1 = np.load("tmp/encode_npy/trial_bald_all.npy")
a2 = np.load("tmp/encode_npy/trial_not_bald_all.npy")

a3 = np.load("tmp/encode_npy/trial_smile_all.npy")
a4 = np.load("tmp/encode_npy/trial_not_smile_all.npy")

a5 = np.load("tmp/encode_npy/trial_smile_MF.npy")
a6 = np.load("tmp/encode_npy/trial_not_smile_MF.npy")
```

```python

# np.save("tmp/encode_npy/trial_smile_100.npy", a5[0:100])
# a5.std()

m1 = np.mean(a1, axis = 0)
m2 = np.mean(a2, axis = 0)
x_bald = m1 - m2

m3 = np.mean(a3, axis = 0)
m4 = np.mean(a4, axis = 0)
x_smile = m3 - m4

m5 = np.mean(a5, axis = 0)
m6 = np.mean(a6, axis = 0)
x_smile_MF = m5 - m6

b = np.zeros((100,64,64,3), dtype = np.float32)

o = np.load("tmp/encode_npy/trial_few.npk")

for i in range(0, 10):
    b[i] = o[3]
    b[i] += x_bald / 5 * i
    
    for j in range(0, 10):
        index = 10*j + i
        b[index] = b[i] - x_smile_MF / 10 * j

np.save("tmp/encode/smile_bald_continuum4.npy", b)
# with open("tmp/encode/bald_continuum_rev.pkl", "w") as f:
#     pickle.dump(b,f)

```
